```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Myntra_sales.csv", encoding = "unicode_escape")
df = pd.DataFrame(data)
print(df)
```

# **DATA CLEANING**

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

# **EDA (Exploratory Data Analysis)**

```python
ax = sns.countplot(data = df, x = "Gender", hue = "Marital_Status", palette = "rocket")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

__Check who can do more shopping Male vs Female__

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
group = df.groupby("Gender")["Amount"].sum()
gp = pd.DataFrame(group)
print(gp)
ax = sns.barplot(data = gp, x = "Gender", hue = "Gender", y = "Amount", palette = "GnBu")
plt.show()
```

__From the above graph we clearly see that Females can do more shopping__

__Count customers on the basis of age group__

```python
ax = sns.countplot(data = df, x = "Age Group", hue = "Gender", palette = "rocket")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

__Check who can do more shopping on the basis of Age-Group__

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
grp = df.groupby("Age Group", as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
grp1 = pd.DataFrame(grp)
print(grp1)
sns.set(rc = {"figure.figsize":(9,4)})
sns.barplot(data = grp1, x= "Age Group", hue = "Age Group", y = "Amount", palette = "viridis", legend = False)
plt.show()
```

__Total numbers of orders state-wise__

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sales_states = df.groupby("State")["Orders"].sum().sort_values(ascending = False) 
grp2 = pd.DataFrame(sales_states)
print(grp2)
sns.set(rc = {"figure.figsize":(16,8)})
ax = sns.barplot(data = grp2, hue = "State", y = "Orders", legend = True, palette = "viridis")
plt.show()
```

__Total amount/sales from top 10 states__

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

amount = df.groupby("State")["Amount"].sum().sort_values(ascending  = False).head(10)
amount_grp = pd.DataFrame(amount)
print(amount_grp)
sns.set(rc = {"figure.figsize":(17,5)})
sns.barplot(data = amount_grp , x = "State", hue = "State" , y = "Amount" , palette = "rocket")
plt.show()
```

__Count the Martial Status__

```python
ax = sns.countplot(data = df , x = "Marital_Status", hue ="Marital_Status",  palette = "GnBu", legend = False)
sns.set(rc = {"figure.figsize":(6,3)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

__Who can do more shopping married/unmarried__

```python
mar = df.groupby(["Marital_Status" , "Gender"])["Amount"].sum()
mar_grp = pd.DataFrame(mar)
print(mar_grp)
sns.set(rc = {"figure.figsize":(5,4)})
sns.barplot(data = mar_grp, x = "Marital_Status", hue = "Gender", y = "Amount" , palette = "Blues")
plt.show()
```

__Number of customers and their job sector__

```python
ax = sns.countplot(data = df , x = "Occupation", hue = "Occupation", palette = "viridis")
sns.set(rc = {"figure.figsize" : (22,6)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

__Product Category__

```python
ax = sns.countplot(data = df, x = "Product_Category" , hue = "Product_Category" , palette = "Accent")
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc ={"figure.figsize" : (27,6)})
plt.show()
```

__Total category/sales__

```python
cat = df.groupby("Product_Category")["Amount"].sum().sort_values(ascending = False).head(10)
cat_grp = pd.DataFrame(cat)
sns.barplot(data = cat_grp , x = "Product_Category" , y = "Amount" , hue ="Product_Category" , palette = "viridis")
plt.show()
```

__Get top selling product__

```python
sell = df.groupby("Product_ID")["Orders"].sum().sort_values(ascending = False).head(12)
sell_grp = pd.DataFrame(sell)
print(sell_grp)
sns.set(rc = {"figure.figsize" : (22,6)})
ax = sns.barplot(data = sell_grp , x = "Product_ID", y = "Orders", hue = "Product_ID", palette = "rocket")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
